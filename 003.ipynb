**19 Сумма кубов и Куб суммы**

На вход программе подаются два целых числа a и b, каждое на новой строке. Ваша программа должна посчитать для этих чисел сумму их кубов и куб их суммы и вывести результат вычислений в следующем формате:

```python
Для чисел <число a> и <число b>:

  Сумма кубов: <число a> ** 3 + <число b> ** 3 = <сумма кубов a и b>

  Куб суммы: (<число a> + <число b>)**3 = <куб суммы a и b>
```

```python
#Входные данные:
1
2

#Вывод программы:
Для чисел 1 и 2:
  Сумма кубов: 1**3 + 2**3 = 9
  Куб суммы: (1 + 2)**3 = 27
```
